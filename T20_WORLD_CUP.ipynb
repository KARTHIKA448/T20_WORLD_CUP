The dataset tells about the T20 World Cup, which is an international cricket tournament in the Twenty20 format.

```python
df.columns
```

each feature:

    venue: The location where the cricket match took place.
    team1: The first team participating in the match.
    team2: The second team participating in the match.
    stage: The stage of the tournament (e.g., Super 12, Semi-final, Final).
    toss winner: The team that won the toss.
    toss decision: The decision made by the toss-winning team (e.g., Field, Bat).
    first innings score: The total score achieved by the team batting first in the first innings.
    first innings wickets: The number of wickets lost by the team batting first in the first innings.
    second innings score: The total score achieved by the team batting second in the second innings.
    second innings wickets: The number of wickets lost by the team batting second in the second innings.
    winner: The winning team of the match.
    won by: The method by which the winning team secured victory (e.g., Runs, Wickets).
    player of the match: The player chosen as the best performer in the match.
    top scorer: The player who scored the most runs in the match.
    highest score: The highest individual score by a batsman in the match.
    best bowler: The player who performed the best in bowling during the match.
    best bowling figure: The best individual bowling performance in terms of figures (e.g., wickets taken and runs conceded)
